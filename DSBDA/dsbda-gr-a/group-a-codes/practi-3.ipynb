```python
# Clean: Drop rows where 'Gender' or 'LoanAmount' is missing
df_clean = df[['Gender', 'LoanAmount']].dropna()
```

```python
# 1. Summary statistics grouped by categorical variable 'Gender'
grouped_stats = df_clean.groupby('Gender')['LoanAmount'].agg(['mean', 'median', 'min', 'max', 'std'])
print("Summary statistics of LoanAmount grouped by Gender:")
print(grouped_stats)
```
